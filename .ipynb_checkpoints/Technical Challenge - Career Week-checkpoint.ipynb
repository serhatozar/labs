```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

df_csv = pd.read_csv('measurements.csv')
df_xlsx = pd.read_excel('measurements2.xlsx')

df_csv
```

```python
df_xlsx
```

```python
df_csv = df_csv.drop_duplicates()
df_csv.reset_index(inplace=True)

df_xlsx = df_xlsx.drop_duplicates()
df_xlsx.reset_index(inplace=True)
```

```python
nulls_df = pd.DataFrame(round(df_csv.isna().sum()/len(df_csv),4)*100)
nulls_df = nulls_df.reset_index()
nulls_df.columns = ['header_name', 'percent_nulls']
nulls_df
```

```python
nulls_df2 = pd.DataFrame(round(df_xlsx.isna().sum()/len(df_xlsx),4)*100)
nulls_df2 = nulls_df2.reset_index()
nulls_df2.columns = ['header_name', 'percent_nulls']
nulls_df2
```

```python
df = df_csv

def lower_case_column_names(columns):
    df.columns = [i.lower() for i in df.columns]
    return df

lower_case_column_names(df)

df.columns = df.columns.str.replace(' ', '_')

df = df.drop(columns=['specials', 'refill_liters', 'refill_gas', 'ac','rain','sun'])

df = df.drop_duplicates()

df['gas_type'] = df['gas_type'].apply(lambda x: 0 if x == 'E10' else 1)

#E10=0 
#SP98=1

df
```

```python
df.dtypes
```

```python
df['distance'] = df['distance'].str.replace(' ', '')
df['consume'] = df['consume'].str.replace(' ', '')
df['temp_inside'] = df['temp_inside'].str.replace(' ', '')
df['distance'] = df['distance'].str.replace(',', '.')
df['consume'] = df['consume'].str.replace(',', '.')
df['temp_inside'] = df['temp_inside'].str.replace(',', '.')
df['distance'] = pd.to_numeric(df['distance'])
df['consume'] = pd.to_numeric(df['consume'])
df['temp_inside'] = pd.to_numeric(df['temp_inside'])
```

```python
df['gas_usage'] = (df['consume'] / 100) * df['distance']


gas_prices = {'0': 1.379, '1': 1.459}

# Calculate the cost per ride based on gas type and distance
df['cost_per_ride'] = (df['consume'] * df['distance'] * df['gas_type'].map(gas_prices))

df['cost_per_ride'] = df['cost_per_ride'].apply(lambda x: f'{x:.4f}')

# Print the DataFrame with the new 'cost_per_ride' column
print(df)
```

```python
correlation_matrix = df[['distance', 'speed', 'consume', 'temp_inside', 'temp_outside']].corr
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 3.1: Bar plot to compare the average fuel consumption for each gas type
average_fuel_consumption = df.groupby('gas_type')['consume'].mean()
plt.figure(figsize=(8, 6))
sns.barplot(x=average_fuel_consumption.index, y=average_fuel_consumption.values)
plt.title('Average Fuel Consumption by Gas Type')
plt.xticks(ticks=[0, 1], labels=['E10', 'SP98'])
plt.xlabel('Gas Type')
plt.ylabel('Average Fuel Consumption (L/100km)')
plt.show()

# 3.2: Scatter plot to visualize the relationship between distance and fuel consumption
plt.figure(figsize=(8, 6))
sns.scatterplot(x='distance', y='consume', data=df, hue='gas_type', palette={0: 'blue', 1: 'red'})
plt.title('Scatter Plot of Distance vs. Fuel Consumption by Gas Type')
plt.xlabel('Distance (km)')
plt.ylabel('Fuel Consumption (L/100km)')
plt.legend(title='Gas Type')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='distance', y='speed', data=df, hue='gas_type', palette={0: 'blue', 1: 'red'})
plt.title('Scatter Plot of Distance vs. Speed')
plt.xlabel('Distance (km)')
plt.ylabel('Speed (km/h)')
plt.legend(title='Gas Type')
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(x='temp_inside', y='temp_outside', data=df, hue='gas_type', palette={0: 'blue', 1: 'red'})
plt.title('Scatter Plot of Tempurature Inside vs. Tempurature Outside')
plt.xlabel('Tempurature Inside (C)')
plt.ylabel('Tempurature Outside (C)')
plt.legend(title='Gas Type')
plt.show()
```
